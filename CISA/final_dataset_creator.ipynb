```python
# import re
# import ast

# def fix_malformed_list_dict_string(s):
#     if not s.strip().startswith('['):
#         print("Skipping non-structured input:", s[:60], "...")
#         return None
#     # Remove outer [[ and ]] brackets
#     s = s.strip()[2:-2]

#     # Add quotes around keys using regex
#     s_fixed = re.sub(r'(\b\w+\b):', r'"\1":', s)

#     # Wrap the result back into a proper list of dicts string
#     s_fixed = '[{' + s_fixed + '}]'

#     # Safely evaluate
#     return s_fixed #ast.literal_eval(s_fixed)

import re
import ast

def fix_malformed_list_dict_string(full_string):
    full_string = full_string.strip()

    # Sanity check
    if not full_string.startswith('[') or ':' not in full_string:
        print("Skipping unstructured input:", full_string[:60], "...")
        return None

    try:
        # Step 1: Split the input into list items
        # Assume each item starts with '{' or '[' and ends with '}' or ']'
        raw_items = re.findall(r'[\[{][^\[\]{}]+[\]}]', full_string)

        fixed_items = []

        for raw in raw_items:
            item = raw.strip()

            # Normalize outer brackets
            item = item.replace('[', '{').replace(']', '}')

            # Add quotes around keys if needed
            item = re.sub(r'([{,]\s*)(\w+)(\s*):', r'\1"\2"\3:', item)

            # Parse as dict
            parsed = ast.literal_eval(item)
            fixed_items.append(parsed)

        return fixed_items

    except Exception as e:
        print("Failed to fully parse input. Error:", e)
        return None




# Examples
s1 = '[[Software: "1756-ENBT EtherNet/IP communication interface", Vulnerability: "Open UDP Port", CVE: "", CWE: ""]]'
s2 = '[[Software: "Microsoft Windows", Vulnerability: ".LNK Vulnerability", CVE: "", CWE: ""]]'

parsed1 = fix_malformed_list_dict_string(s1)
parsed2 = fix_malformed_list_dict_string(s2)
```
