````python
#Embedding based match

# pip install sentence-transformers faiss-cpu tqdm
from sentence_transformers import SentenceTransformer
import faiss, numpy as np
from collections import defaultdict
from tqdm import tqdm

# ── 1. Load model ────────────────────────────────────────────────────────────
MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
model      = SentenceTransformer(MODEL_NAME)

# ── 2. Pre-compute CVE embeddings once ───────────────────────────────────────
cve_ids, cve_texts = [], []
for cve in cves:
    cve_ids.append(cve["identifier"])
    cve_texts.append(cve["description"]["description_data"][0]["value"])

cve_embs = model.encode(
    cve_texts, 
    convert_to_numpy=True, 
    normalize_embeddings=True, # → cosine sim == dot product
    show_progress_bar=True 
)

d      = cve_embs.shape[1]        # embedding dimension (384)
index  = faiss.IndexFlatIP(d)     # inner-product ≡ cosine for L2-norm 1 vectors
index.add(cve_embs)

# ── 3. Search parameters ─────────────────────────────────────────────────────
THRESH   = 0.75                   # cosine sim threshold for a “match”
TOP_K    = 50                     # retrieve up to K nearest neighbours

# ── 4. Outputs & cache ───────────────────────────────────────────────────────
match_counts = defaultdict(list)  # needle → [CVE IDs] | ["No Match"]
cache        = {}                 # lower-cased needle → same list

# ── 5. Process relations ────────────────────────────────────────────────────
for item in tqdm(overal_relations):
    vals = item.split(", ")
    if len(vals) < 2:
        continue

    second = vals[1]
    if second.startswith("CVE") or "vulnerabilities" in second:
        continue

    needle_raw = second
    needle_lc  = needle_raw.lower()

    # cached?
    if needle_lc in cache:
        match_counts[needle_raw].extend(cache[needle_lc])
        continue

    # embed query once
    q_emb = model.encode([needle_raw], convert_to_numpy=True, normalize_embeddings=True)
    sims, idxs = index.search(q_emb, TOP_K)          # sims: shape (1, TOP_K)

    hits = [cve_ids[i] for sim, i in zip(sims[0], idxs[0]) if sim >= THRESH]

    cache[needle_lc] = hits if hits else ["No Match"]
    match_counts[needle_raw].extend(cache[needle_lc])


````

````python
import re, ast, json
from typing import Dict, List
from openai import OpenAI

# ──────────────────────────────────────────────────────────────────── #
# Helper: progressively try to "heal" common JSON problems            #
# ──────────────────────────────────────────────────────────────────── #

def _safe_json_parse(raw: str) -> Dict[str, List[str]]:
    """
    Attempt to coerce the model's output into a proper dict.
    Raises ValueError only if all heuristics fail.
    """
    txt = raw.strip()

    # 1) Strip ```json … ``` or ``` … ``` fences
    m = re.search(r"```(?:json)?\s*(\{.*?\})\s*```", txt, flags=re.DOTALL)
    if m:
        txt = m.group(1).strip()

    # 2) First try a straight json.loads
    try:
        return json.loads(txt)
    except json.JSONDecodeError:
        pass

    # 3) Remove trailing commas before } or ]
    txt2 = re.sub(r",\s*([}\]])", r"\1", txt)
    try:
        return json.loads(txt2)
    except json.JSONDecodeError:
        pass

    # 4) Replace single quotes → double quotes (quick-n-dirty)
    txt3 = txt2.replace("'", '"')
    try:
        return json.loads(txt3)
    except json.JSONDecodeError:
        pass

    # 5) Last-ditch: ast.literal_eval (handles Python-ish dicts)
    try:
        return ast.literal_eval(txt3)
    except Exception as exc:
        raise ValueError(f"Unable to parse model output as JSON:\n{raw}") from exc
````

````python
SYSTEM_PROMPT = (
    "You are a veteran software/hardware security analyst who maintains a "
    "knowledge-graph of real-world vulnerabilities. "
    "Your job is to decide which candidate names below refer to *distinct, "
    "specific vulnerabilities tied to a concrete software or hardware product*. "
    "Reject generic terms like ‘elevation of privilege’. "
    "Return strictly-formatted JSON with two keys: "
    "specific_vulnerabilities and too_general."
)

RESPONSE_SCHEMA_EXAMPLE = {
    "specific_vulnerabilities": ["CVE-2021-44228", "Cisco IOS XR BGP DoS (CSCvx69814)"],
    "too_general": ["elevation of privilege", "side-channel attack"],
}

def chunk_list(seq: List[str], size: int) -> Generator[List[str], None, None]:
    for i in range(0, len(seq), size):
        yield seq[i : i + size]

def build_prompt(chunk: List[str]) -> str:
    chunk_pretty = json.dumps(chunk, indent=2)
    example_pretty = json.dumps(RESPONSE_SCHEMA_EXAMPLE, indent=2)
    return (
        f"Below is a list of vulnerability *names*:\n\n{chunk_pretty}\n\n"
        "TASK\n"
        "----\n"
        "Identify which names are concrete vulnerabilities tied to a specific "
        "product and which are too general. Remove duplicates (case-insensitive). "
        "Respond **only** with JSON exactly in this schema:\n\n"
        f"{example_pretty}\n\n"
        "No extra commentary."
    )

def call_llm(client: OpenAI, model: str, prompt: str) -> Dict[str, List[str]]:
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": prompt},
        ],
        temperature=0.0,
    )
    content = response.choices[0].message.content.strip()
    
        # robust parsing
    parsed = _safe_json_parse(content)
    
    parsed["specific_vulnerabilities"] = [v.strip() for v in parsed["specific_vulnerabilities"]]
    parsed["too_general"] = [v.strip() for v in parsed["too_general"]]
    return parsed
````

````python
# ─── user-editable parameters ──────────────────────────────────────────
out_dir     = "/nas02/Hadi/NER-New/New_Relation_Extraction/verified_vulns"       # where to save results
batch_size  = 100                          # ~100 keeps context safe
model_name  = "gpt-4o"
request_delay = 0.5                        # seconds between calls
# ──────────────────────────────────────────────────────────────────────

#api_key = os.getenv("OPENAI_API_KEY")      # or set directly:
# api_key = "sk-..."                       # <- uncomment if you prefer 
api_key =''
if not api_key:
    raise EnvironmentError("Set OPENAI_API_KEY env-var or hard-code api_key.")

client = OpenAI(api_key=api_key)
````

````python
from typing import List, Dict, Any

def extract_keys(vuln_dict: Dict[Any, Any]) -> List[str]:
    """
    Flatten a dictionary into a clean list of candidate vulnerability names.
    
    Handles three common patterns:
      • {key: "vuln_name"}
      • {key: ["vuln1", "vuln2", ...]}
      • {key: {"name": "vuln_name", ...}}  (keep the 'name' value)
    Modify the inner logic if your structure is different.
    """
    collected = []
    for key, val in vuln_dict.items():
        # 1) Simple string
        if isinstance(key, str):
            collected.append(key)
        # 2) Iterable of strings
        elif isinstance(key, (list, tuple, set)):
            collected.extend(str(x) for x in key)
        # 3) Nested dict with a 'name' field
        elif isinstance(key, dict) and "name" in key:
            collected.append(key["name"])
        else:
            # fall back to string-ify unknown structures
            collected.append(str(key))
    
    # Strip whitespace, drop empties, deduplicate while preserving order
    seen = set()
    cleaned = []
    for v in collected:
        v = v.strip()
        if v and v.lower() not in seen:
            seen.add(v.lower())
            cleaned.append(v)
    return cleaned

# ------ use it --------------------------------------------------------

keys = extract_keys(match_counts)
print(f"Loaded {len(keys)} vulnerability names from in-memory dictionary.")
````

````python
specific_all, general_all = [], []

for chunk in tqdm(list(chunk_list(keys, batch_size)), desc="Verifying"):
    prompt = build_prompt(chunk)
    result = call_llm(client, model_name, prompt)
    specific_all.extend(result["specific_vulnerabilities"])
    general_all.extend(result["too_general"])
    time.sleep(request_delay)

# case-insensitive de-duplication + sort
specific_all = sorted(set(v.lower() for v in specific_all))
general_all  = sorted(set(v.lower() for v in general_all))

print(f"✅ {len(specific_all)} specific vulnerabilities")
print(f"❌ {len(general_all)} too general/unclear")
````

````python
def remove_between_tags(text):
    # Remove everything between < and > including the symbols
    text = re.sub(r'<.*?>', '', text)
    # Remove any extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text
````
